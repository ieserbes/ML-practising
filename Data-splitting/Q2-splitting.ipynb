1.	Load the data 

```python
#Import the necessary libraries
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.model_selection import train_test_split

#Load the "mysteryDataSplit" data and assign to a dataframe called df
df = pd.read_csv('mysteryDataSplit.csv')
#check the data if imported correctly (visual inspection)
df
```

    * The data was imported correctly. Visual inspection indicates that we have 1000 samples, 20 inputs (features) and one target ("Class"), suggesting that in hypothetical scenerio, this data would be using by supervised ML algorithm for a classification problem. 

2.	Generate a split set with random class proportions ready for ML

    * In the question, it was asked to "ignore validation dataset", I therefore will only split the data for train/test sets. 
    (The reason that we ignore the validation dataset in hypothetical scenerio, this data set would be using to train only one model rather than multiple models and therefore no need for validation.)
    
    * We have 1000 samples and therefore we can tell that this is considerably a large data set. 
    * Further more, since we exclude the validation data, we can use more data for model training (in the module, the recommended data split was 50% training, 25% validation and 25% test), which allows the model to learn more complex patterns and improve its performance (providing more sample for training might indeed necessary since we have currently 20 inputs (if we would not go for feature selection)). 
    * For this reason (and we also learned in the module Unit1), it is a common practise to split data as 80% training and 20% test (or 70% training and 30%test). Since we have relavitely large dataset, I decided to divide the data as 80% training (to train the model) and 20% test (to test the generalization of the model). With this way, I will have sufficient data to train the model and reasonable amount of data to evaluate the model's performance on unseen samples.


```python
#Data split

# assign the inputs as X and target("Class") as y 

X = df.iloc[:, 0:20] #subset of first 20 columns from df
y = df.iloc[:, 20]   #subset the class column from a series

#import the IPython.display for visualization the dataframe more structured
from IPython.display import display

display(X.head(5))
display(y.head(5))

# generate data split for train and test data set with random class proportions
x_train, x_test, y_train, y_test =train_test_split(X, y, test_size=0.2, random_state=63)

#visual inspection

##for x_train
print(f"x_train contains {x_train.shape[0]} samples and {x_train.shape[1]} features.")
display (x_train.head(5))

##for x_test
print(f"x_test contains {x_test.shape[0]} samples and {x_test.shape[1]} features.")
display (x_test.head(5))

##for y_train
print(f"y_train contains {y_train.shape[0]} samples.")
print(y_train.head(5))

##for y_test
print(f"y_test contains {y_test.shape[0]} samples.")
print(y_test.head(5))
```

3. Visualize class assignment in each part of your split

```python
#Count classes in each

## I previously assigned the Class column in df into y variable and then divided to test and train randomly

# count classes in the training set (y_train) and convert the data series into data frame
class_counts_train = y_train.value_counts()
#visualize the result
print("class count in training set")
print(class_counts_train.sort_index())

# count classes in the test set (y_test) 
class_counts_test = y_test.value_counts()
#visualize the result
print("class count in test set")
print(class_counts_test.sort_index())
```

```python
#plot

#import seaborn
import seaborn as sns

# display the two plots as two subplots in one figure
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Plot class counts in the training set
sns.countplot(x=y_train, ax=axes[0]) #set the plot and its subplot position
axes[0].set_title("Class Distribution in Training Set") #add the title of the plot
axes[0].set_xlabel("Class") #add the x axis title 
axes[0].set_ylabel("Count") #add the y axis title 

# Plot class counts in the test set
sns.countplot(x=y_test, ax=axes[1]) #set the plot and its subplot position
axes[1].set_title("Class Distribution in Test Set") #add the title of the plot
axes[1].set_xlabel("Class") #add the x axis title 
axes[1].set_ylabel("Count")#add the y axis title 

# Adjust spacing between subplots
plt.tight_layout()
```

    * The plots indicates that the proportions of class distributions per each data sets are not preserved in train and test data due to randomness.
    

```python
#you may also want to flip the df and plot

"""we can flip the data frame by transpose function""" 
""" but y train and y test are in data series, we can first convert them into data frame and transpose"""
##for y_train
df_train = pd.DataFrame({'Class': y_train}) 
df_train_flip=df_train.T

#visualize the flipped dataframe
display("df_train_flip")
display(df_train_flip)

##for y_test
df_test = pd.DataFrame({'Class': y_test}) 
df_test_flip=df_test.T
#visualize the flipped dataframe
display("df_test_flip")
display(df_test_flip)

#In seaborn, we actually don't have to flip the data but we can assign it to as y axis rather than x axis.
#Therefore we don't need to use transposed df_train or df_test
# display the two plots as two subplots in one figure (flipped)
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Plot class counts in the training set (flipped)
sns.countplot(y=y_train, ax=axes[0]) #set the plot and its subplot position
axes[0].set_title("Class Distribution in Training Set") #add the title of the plot
axes[0].set_xlabel("Class") #add the x axis title 
axes[0].set_ylabel("Count") #add the y axis title 

# Plot class counts in the test set (flipped)
sns.countplot(y=y_test, ax=axes[1]) #set the plot and its subplot position
axes[1].set_title("Class Distribution in Test Set") #add the title of the plot
axes[1].set_xlabel("Class") #add the x axis title 
axes[1].set_ylabel("Count")#add the y axis title 

# Adjust spacing between subplots
plt.tight_layout()
```

```python
#Or by proportion (each count as a % of total in the split)

#Previously we count the classes :"class_counts_train" and "class_counts_test"
#now we can divide them from total and multiply with 100

#for train
class_percentages_train = (class_counts_train / len(y_train)) * 100

#for test
class_percentages_test = (class_counts_test / len(y_test)) * 100

#we can plot them

# display the two plots as two subplots in one figure
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Plot class percentages in the training set
sns.barplot(x=class_percentages_train.index, y=class_percentages_train.values, ax=axes[0])
axes[0].set_title("Class Distribution in Training Set")
axes[0].set_xlabel("Class")
axes[0].set_ylabel("%")

# Plot class percentages in the test set
sns.barplot(x=class_percentages_test.index, y=class_percentages_test.values, ax=axes[1])
axes[1].set_title("Class Distribution in Test Set")
axes[1].set_xlabel("Class")
axes[1].set_ylabel("%")

# Adjust spacing between subplots
plt.tight_layout()

```
